```python
import pandas as pd 
```

```python
data = pd.read_excel(io = "Performance Dashboard.xlsx", engine = "openpyxl",  skiprows=[0])
```

```python
data.head()
```

```python
nan_value = float("NaN") 

data.replace("", nan_value, inplace=True) 


data.dropna(subset = ["County"], inplace=True) 

data.head(10)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head(10)
```

```python
data.columns.values.tolist()
```

```python
new_data = data.rename(columns={
                      "MonthTarget" : "CIP_Month Target ", 
                      "MTD Target": "CIP_MTD Target", 
                      "MTD Actual" : "CIP_MTD Actual",
                     "MTD Performance" : "CIP_MTD Performance",
                     "MonthTarget.1" : "CR_MonthTarget",
                     "MTD Target.1" : "CR_MTD Target",
                     "MTD Actual.1" : "CR_MTD Actual", 
                     "MTD Performance.1" : "CR_MTD Performance",
    
                     "MonthTarget.2" : "MKTA_MonthTarget",
                     "MTD Target.2" : "MKTA_MTD Target",
                     "MTD Actual.2" : "MKTA_MTD Actual",
                     "MTD Performance.2" : "MKTA_MTD Performance",
    
                     "MTD Actual.3" : "Total_MTD Actual",
                     "MTD Performance.3" : "Total_MTD Performance"
                    })
            
```

```python
new_data.head()
```

```python
import plotly.express as px 
import streamlit as st
```

```python
writer = pd.ExcelWriter('Clean Performance Dashboard.xlsx')
```

```python
new_data.to_excel(writer)
```

```python
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
